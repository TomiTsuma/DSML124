```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from scipy.spatial.distance import mahalanobis
from sklearn.preprocessing import StandardScaler
```

```python
pcc1_original = pd.read_csv("/home/azureuser/cloudfiles/code/Users/tsuma.thomas/DSML124/outputFiles/PCC1/test/aluminium.csv",index_col=0)
pcc1_reconstructed = pd.read_csv("/home/azureuser/cloudfiles/code/Users/tsuma.thomas/DSML124/outputFiles/pcc1_reconstructed_spc/aluminium.csv",index_col=0)

```

```python
pcc2_original = pd.read_csv("/home/azureuser/cloudfiles/code/Users/tsuma.thomas/DSML124/outputFiles/PCC2/aluminium.csv",index_col=0)
pcc2_reconstructed = pd.read_csv("/home/azureuser/cloudfiles/code/Users/tsuma.thomas/DSML124/outputFiles/pcc2_reconstructed_spc/aluminium.csv",index_col=0)

```

```python
pcc1_reconstructed.columns = pcc1_original.columns
```

```python
pcc2_reconstructed.columns = pcc2_original.columns
```

```python
mahalanobis_distance_pcc1 = []
for i in range(len(pcc1_original)):
    df = pd.concat([pd.DataFrame(pcc1_original.iloc[i]).T, pd.DataFrame(pcc1_reconstructed.iloc[i]).T])
    X = np.array(df)  # Example dataset containing u and v
    cov_matrix = np.cov(X.T)
    inv_cov_matrix = np.linalg.inv(cov_matrix)
    distance = mahalanobis(pd.DataFrame(pcc1_original.iloc[i]).T.values[0], pd.DataFrame(pcc1_reconstructed.iloc[i]).T.values[0], inv_cov_matrix)
    mahalanobis_distance_pcc1.append(distance)    
```

```python
mahalanobis_distance_pcc2 = []
for i in range(len(pcc2_original)):
    df = pd.concat([pd.DataFrame(pcc2_original.iloc[i]).T, pd.DataFrame(pcc2_reconstructed.iloc[i]).T])
    X = np.array(df)  # Example dataset containing u and v
    cov_matrix = np.cov(X.T)
    inv_cov_matrix = np.linalg.inv(cov_matrix)
    print(np.shape(inv_cov_matrix))
    distance = mahalanobis(pd.DataFrame(pcc2_original.iloc[i]).T.values[0], pd.DataFrame(pcc2_reconstructed.iloc[i]).T.values[0], inv_cov_matrix)
    mahalanobis_distance_pcc2.append(distance)
    
```

```python
pcc1_original['mahalanobis_distance'] = mahalanobis_distance_pcc1
pcc2_original['mahalanobis_distance'] = mahalanobis_distance_pcc2
```
