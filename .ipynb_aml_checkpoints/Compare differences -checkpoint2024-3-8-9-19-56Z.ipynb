```python
chemicals = [
    'aluminium',
    'phosphorus',
    'ph',
    'exchangeable_acidity',
     'calcium',
    'magnesium',
    'sulphur', 
    'sodium',
    'iron', 
    'manganese',
    'boron', 
    'copper', 
    'zinc', 
    'total_nitrogen', 
    'potassium',
    'ec_salts', 
    'organic_carbon', 'cec',
     'sand'
      ,'silt', 'clay'
]
```

```python
df = pd.DataFrame()
difference_thresholds = {}
for chem in chemicals:
    df_pcc1 = pd.read_csv(f"outputFiles/PCC1_Classes_Reconstructed/{chem}.csv")
    df_pcc2 = pd.read_csv(f"outputFiles/PCC2_Classes_Reconstructed/{chem}.csv")
    thresholds = pd.DataFrame(df_pcc1.Difference.describe().T)
    
    thresholds = thresholds.drop('count')
    thresholds = thresholds.drop('std')
    metrics = []
    tps = []
    tns = []
    fps = []
    fns = []
    percentiles = [i for i in (np.arange(10, 100, 10))]
    
    
    for value in percentiles:
        metrics.append(value)
        val = value / 100
        tps.append(len(df_pcc2.loc[df_pcc2['Difference'] >= np.quantile(df_pcc1['Difference'], val)]))
        tns.append(len(df_pcc1.loc[df_pcc1['Difference'] < np.quantile(df_pcc1['Difference'], val)]))
        fps.append(len(df_pcc1.loc[df_pcc1['Difference'] >= np.quantile(df_pcc1['Difference'], val)]))
        fns.append(len(df_pcc2.loc[df_pcc2['Difference'] < np.quantile(df_pcc1['Difference'], val)]))
#     chem_
    chem_df = pd.DataFrame({"chemical":chem, "percentile":metrics,"TP": tps,"TN": tns,"FP": fps,"FN": fns, "outliers":len(df_pcc2), "normal":len(df_pcc1)})
    df = pd.concat([df, chem_df])
```

```python
df.to_csv("DecisionMatrix.csv")
```

```python
precisions = []
recalls = []
f1s = []
specificities = []
for index, row in df.iterrows():
    print(row)
    fp = row['FP']
    fn = row['FN']
    tp = row['TP']
    tn = row['TN']

    if(tp == 0):
        recalls.append(0)
        precisions.append(0)
        f1s.append(0)
        specificities.append(0)
        continue

    print(row['chemical'])
    # print(row['percentile'])

    denom = (tp+fn)
    numer = tp
    recall = (numer / denom)
    denom = (tp+fp)
    numer = tp
    
    precision = (numer / denom)

    print("...........................")
    print(tp)
    print(tp+fn)
    print(tp+fp)
    # print(recall)
    # print(precision)
    print("...........................")
    
    f1_score = ((2 * recall * precision) / (recall + precision))
    specificity =  (tn / (tn+fp))

    recalls.append(recall)
    precisions.append(precision)
    f1s.append(f1_score)
    specificities.append(specificity)

df['precision'] = precisions
df['recall'] = recalls
df['f1_score'] = f1s
df['specificity'] = specificities

  
```

```python
df.to_csv("DecisionMatrix.csv")
```

```python
thresholds = {}
for chem in chemicals:
    df_pcc1 = pd.read_csv(f"outputFiles/PCC1_Classes_Reconstructed/{chem}.csv")
    _ = df.loc[df['chemical'] == chem]
    max_f1 = _['f1_score'].max()
    percentile = _.loc[_['f1_score'] == max_f1]['percentile'].values[0]
    print(type(percentile))
    threshold = np.quantile(df_pcc1['Difference'], percentile/100)
    thresholds[chem] = threshold
```

```python
def true_positive(metric ,threshold, count):
    print(metric)
    print(threshold)
    print(count)
    percent_correctly_classified = len(df_pcc2.loc[df_pcc2['Difference'] > threshold]) / len (df_pcc2)
    print(f"Percentage of correct true positive: {percent_correctly_classified}")
    count_correctly_classified = percent_correctly_classified * count
    print(f"Number of correctly classified: {count_correctly_classified}")
    
    cost_of_tp = 100 * count_correctly_classified
    print(f"Cost of tp: {cost_of_tp}")
    
    count_wrongly_classified = count - count_correctly_classified
    print(f"Number of wrongly classified: {count_wrongly_classified}")
    cost_of_fp = 200 * count_wrongly_classified
    print(f"Cost of fp: {cost_of_fp}")
```
